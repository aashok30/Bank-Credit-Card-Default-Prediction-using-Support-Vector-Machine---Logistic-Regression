```python
#Droping the Customer ID as it is a unique column.
df= df.drop('Customer ID',axis=1)
```

```python
x=df.drop('Default_Payment',axis=1)
y=df['Default_Payment']
```

```python
#Performing train_test_split.
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3, random_state=0)
```

```python
#Scaling the data.
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
sc.fit_transform(x_train)
```

```python
x_train=sc.transform(x_train)
```

```python
x_test=sc.transform(x_test)
```

```python
# Build Support vector machine model 
from sklearn.svm import SVC
svc=SVC(random_state = 42) # default parameters
```

```python
svc.fit(x_train, y_train)
```

```python
y_pred = svc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_curve, roc_auc_score
print("Accuracy of Model is: ",accuracy_score(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.metrics import plot_confusion_matrix, plot_roc_curve, plot_precision_recall_curve
plot_confusion_matrix(svc, x_test, y_test, values_format='d', display_labels=['Non - Default', 'Defaulted'])
```

Analyzing Confusion Matrix:

Accuracy = (TP+TN)/(TP+FP+FN+TN)
        = (6707+712)/(6707+1244+321+712)
        = (7419)/(8984)
        = 83%

Sensitivity/Recall = TP/(TP+FN)
                    = 6707/(6707+321)
                    = 6707/7028
                    = 95%



Specificity = TN/(TN+FP)
            = 712/(712+1244)
            = 712/1956
            = 36%

F1 Score = 2*(Recall * Precision) / (Recall + Precision)
            = 2*(0.95*0.84)/(0.95+0.84)
            = 2*(0.798)/(1.79)
            = 2*0.45
            = 90%

From 6707 + 321 = 7028 people that Did Not Default, 321 was missclassified (5%)

From 1244 + 712 = 1956 people that Defaulted, 712 was missclassified (36%)

That is not acceptable!
